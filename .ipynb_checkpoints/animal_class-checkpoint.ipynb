```python
def convert_binary(x):
    if x == 'Yes':
        return 1
    if x == 'No':
        return 0
```

```python
df3=pd.DataFrame()
```

```python
df3['hair']=df2['hair'].apply(convert_binary)
df3['feathers']=df2['feathers'].apply(convert_binary)
df3['eggs']=df2['eggs'].apply(convert_binary)
df3['milk']=df2['milk'].apply(convert_binary)
df3['airborne']=df2['airborne'].apply(convert_binary)
df3['aquatic']=df2['aquatic'].apply(convert_binary)
df3['predator']=df2['predator'].apply(convert_binary)
df3['toothed']=df2['toothed'].apply(convert_binary)
df3['backbone']=df2['backbone'].apply(convert_binary)
df3['breathes']=df2['breathes'].apply(convert_binary)
df3['venomous']=df2['venomous'].apply(convert_binary)
df3['fins']=df2['fins'].apply(convert_binary)
df3['tail']=df2['tail'].apply(convert_binary)
df3['domestic']=df2['domestic'].apply(convert_binary)
df3['catsize']=df2['catsize'].apply(convert_binary)

```

```python
y=df2.drop(['hair','feathers','eggs','milk','airborne','aquatic','predator','toothed','backbone','breathes','venomous','fins',
           'legs','tail','domestic','catsize'],axis='columns')
```

```python
y
```

```python
y.class_type.unique()
```

```python
def class_binary(x):
    if x == 'Mammal':
        return 0
    if x == 'Fish':
        return 1
    if x == 'Bird':
        return 2
    if x == 'Invertebrate':
        return 3
    if x == 'Bug':
        return 4
    if x == 'Amphibian':
        return 5
    if x == 'Reptile':
        return 6
```

```python
y['class_type_n']=y['class_type'].apply(class_binary)
```

```python
y=y.drop(['class_type'],axis='columns')
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
m = LogisticRegression(solver='lbfgs',class_weight='balanced', max_iter=20000)
```

```python
m.fit(X_train,y_train.values.ravel())
```

```python
m.score(X_test,y_test)
```
